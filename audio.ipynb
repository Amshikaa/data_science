```python
import librosa
import IPython.display as ipd

file_path='sample.wav'
audio_data,sample_rate=librosa.load(file_path,sr=None)

audio_weight=ipd.Audio(data=audio_data,rate=sample_rate)

display(audio_weight)
```

```python
#without using builtin function
def read_mp3_file(file_path):
    with open(file_path, 'rb') as mp3_file:
        mp3_file.seek(100) 
        audio_data = mp3_file.read()
        return audio_data
audio_path = 'sample.wav' 
mp3_audio_data = read_mp3_file(audio_path)

print("MP3 Audio Data Length:", mp3_audio_data)
print(mp3_audio_data)
```

```python
def read_wav_audio(file_path):
    with open(file_path, 'rb') as wav_file:
        wav_file.seek(44) 
        audio_data = wav_file.read()
    return audio_data
audio_path = 'sample.wav'

# Read the WAV file
wav_audio_data = read_wav_audio(audio_path)

# Print a portion of the WAV audio data (as raw binary data)
print("WAV Audio Data (Partial):", wav_audio_data[:10])  # Print the first 10 bytes as an example
```
